## 1. Modelo de perceptrón

La neurona no es más que un conjunto de entradas, pesos y sesgos (comúnmente denominados x, ω y b ), y una función de activación. El proceso para obtener la salida, consiste en que a todas las entradas recibidas, se les multiplica su peso asociado, posteriormente se hace una suma ponderada de los mismos, añadiendo el sesgo, y al resultado de dicho cálculo se le aplica la función de activación, generando de esa manera el resultado final.

Una red neuronal simple, también llamada perceptrón, consta de 2 capas, la de entrada y la de la salida, aunque algunos textos se refieren a esta red neuronal como de 1 capa, pues la capa de entrada solo recibe los datos, y la de la salida es la que realmente procesa la información y la entrega. (Izadkhah, 2022).

Visto lo anterior, podemos decir que el perceptrón se puede definir como una función matemática ""f" que permite relacionar un vector de variables independientes "x" = (x1,… ,xN) , con una variable dependiente "y" , de manera que se estima la variable independiente como ŷ = f(x) . Esta función está parametrizada por un vector de pesos, ω = (ω1,…,ωN)^T , que ponderan la influencia de cada una de las variables independientes en el valor de la variable dependiente, y por una función φ, llamada la función de activación , de manera qu: ŷ = φ(ω*x) (Garau, 2021b).

* ¿Cómo se usa en la regresión lineal?

Para poder realizar regresión lineal con un perceptrón, es necesario contar con una variable independiente (x) y una variable dependiente (y). Ahora bien, debido a que nos interesa estimar un valor continuo y lineal, no requerimos de usar ninguna función de activación, pues estás son usadas más para valores discretos y no lineales. Pues bien, una vez aclarado esto, tendríamos como resultado el siguiente modelo:: ŷ = ωx + b, el cuál se asemeja a la ecuación que hemos visto en la escuela de una línea recta: y = mx + b. Y de esa forma, con este modelo y entrenándolo de manera apropiada, nos dará una regresión lineal (BaronApe3656, 2023).

* ¿Cómo se usa para clasificación?

En el caso de clasificación binaria, podemos tomar diferentes entradas "x", ya que la naturaleza entre las variables independientes y dependiente no es lineal. Los datos de entrada representan característica que nos interesa estudiar y clasificar, pero las características pueden ser muy diferentes entre las diferentes aplicaciones que nos interese obtener algún resultado de tipo binario, ya sea para diagnosticar si un paciente tiene una enfermedad o no, diferenciar imágenes de perros y gatos, o según las calificaciones escolares de un alumno predecir si pasa la materia al final del semestre o no. Pues bien, al ser un problema más complejo que la regresión lineal, necesitamos que nuestro modelo aprenda de estas características y encuentre alguna división entre los datos que nos ayude a clasificarlos.

En clasificación binaria, si es necesario usar una función de activación, puesto que el calculo de la estimación suelen ser valores decimales, los cuales no nos dice con exactidud a que tipo de clase pertenece (0 o 1). La función de activación más común que se suele implementar para este tipo de modelo es la función sigmoidal. Si el valor estimado es igual o mayor a cierto umbral, siendo 0.5 en clasificación binaria, se toma una salida con valor de 1, pero si la estimación es menor a 0.5, la salida toma un valor de 0, y de esta manera se hace la clasificación binaria.  (Garau, 2021a).

2. Funciones de activación para clasificación: Sigmoide y Tangente Hiperbólica (Tanh)

Cuando trabajas con clasificación, el objetivo es asignar cada entrada a una categoría (generalmente 2 en el caso de la clasificación binaria). Aquí es donde las funciones no lineales como la sigmoide y tangente hiperbólica son útiles, ya que ayudan a transformar la salida en un rango limitado que se puede interpretar como una probabilidad.

a) Función sigmoide

- ¿Qué es?
La función sigmoide convierte cualquier valor de entrada en un número entre 0 y 1, lo que la hace ideal para tareas de clasificación binaria.

Aquí e es la constante de Euler (~2.718). Cuando la entrada es muy grande, el valor se aproxima a 1, y cuando es muy pequeña (negativa), se aproxima a 0.

- ¿Por qué se usa en clasificación?:
Se utiliza porque su salida es fácil de interpretar como una probabilidad. Como por ejemplo, si el resultado es cercano a 1, podemos decir que la entrada pertenece a una clase, y si es cercano a 0, pertenece a otra.


b) Función tangente hiperbólica (Tanh)
- ¿Qué es?:
 La función Tanh es muy parecida a la sigmoide, pero su salida está en el rango de -1 a 1 en lugar de 0 a 1.

- Fórmula:

f(x)=tanh(x)=ex-e-xex+e-x

Esto significa lo previamente comentado. que los valores de salida varían entre -1 y 1, donde los valores positivos indican una clase y los negativos otra.
- ¿Por qué se usa en clasificación?:
Se utiliza en problemas de clasificación donde es útil trabajar con salidas tanto positivas como negativas. Además, su forma "S" (similar a la sigmoide) ayuda a separar los datos en dos clases distintas.

# **Caso de uso en el mundo real: Clasificación de mensajes de texto como spam**
Uno de los usos más comunes del modelo de perceptrón es la clasificación binaria, como la clasificación de correos electrónicos o mensajes de texto en dos categorías: spam y no spam. Este problema se puede abordar utilizando un perceptrón de una sola neurona, donde las características de entrada son características extraídas del contenido de los mensajes y la salida es una predicción sobre si el mensaje es spam (1) o no spam (0).


A continuccion en el bloque se importan las librerias necesarias para trabajar en este proyecto, asi como Se verifica y descarga el conjunto de palabras vacías (stopwords) de NLTK si aún no está disponible. Las stopwords son palabras comunes que generalmente se eliminan en el preprocesamiento de texto, ya que no aportan información significativa para la clasificación.

```python
# Importamos las bibliotecas necesarias
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

# Establecemos una semilla global para la reproducibilidad
tf.random.set_seed(4500)

# Descargamos stopwords si aún no están descargadas
import nltk
nltk.download('stopwords', quiet=True)

# Cargamos el conjunto de datos
data = pd.read_csv('Spam_SMS.csv')  # Asegúrate de que el nombre del archivo coincida con el subido
```

Analizando la informacion obtenida podemos decir que: Ambas columnas (Class y Message) tienen 5574 valores no nulos, lo que significa que no hay datos faltantes en este conjunto.

Ademas ambas columnas son de tipo object, lo que indica que contienen datos de texto.
Hay una diferencia notable en la cantidad de mensajes etiquetados como ham y spam. En total, hay 5574 mensajes, de los cuales el 86.6% (4827) son ham y el 13.4% (747) son spam.

Esta distribución muestra que el conjunto de datos está desbalanceado, con una cantidad considerablemente mayor de mensajes ham en comparación con spam. Esto puede ser problemático para el modelo, ya que puede tender a predecir más la clase mayoritaria (ham) y no aprender adecuadamente a identificar la clase minoritaria (spam).

```python

# 1. Descripción del Conjunto de Datos y análisis Exploratorio de Datos
print("Información del Conjunto de Datos:")
data.info()  # Muestra información básica sobre el dataframe
print("\nDistribución de Clases:")
print(data['Class'].value_counts())  # Muestra la cantidad de instancias de cada clase

# Visualizamos la distribución de clases
sns.countplot(x='Class', data=data)  # Gráfico de conteo para visualizar la distribución
plt.title("Distribución de Clases")
plt.xticks(ticks=[0, 1], labels=['Ham', 'Spam'])  # Etiquetas en el eje x
plt.show()

```

La función preprocess_text es esencial para preparar los textos antes de ser procesados por el modelo. Realiza varias tareas clave: convierte todo el texto a minúsculas para evitar distinciones entre mayúsculas y minúsculas, elimina caracteres no alfabéticos para enfocarse en el contenido relevante, y divide el texto en tokens, que son palabras individuales. Además, elimina las stopwords palabras comunes que no aportan valor y aplica stemming, reduciendo las palabras a su raíz, quitando prefijos y sufijos. Este preprocesamiento asegura que los datos de entrada sean limpios y estandarizados, lo que mejora la capacidad del modelo para aprender patrones significativos y realizar predicciones precisas.

```python
# 2. Data Preparation
# Mapeamos 'ham' a 0 y 'spam' a 1
data['Class'] = data['Class'].map({'ham': 0, 'spam': 1})

# Función de preprocesamiento del texto
def preprocess_text(text):
    text = text.lower()  # Convertimos el texto a minúsculas
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Eliminamos caracteres no alfabéticos
    tokens = text.split()  # Dividimos el texto en tokens
    stop_words = set(stopwords.words('english'))  # Obtenemos las stopwords en inglés
    ps = PorterStemmer()  # Inicializamos el stemming
    tokens = [ps.stem(word) for word in tokens if word not in stop_words]  # Stemming y eliminación de stopwords
    return ' '.join(tokens)  # Devolvemos el texto procesado

# Aplicamos el preprocesamiento a los datos
data['Processed_Message'] = data['Message'].apply(preprocess_text)  # Creamos una nueva columna con los mensajes procesados


```

Se separan los mensajes procesados (X) y las etiquetas de clase (y). Ademas, se utiliza el TfidfVectorizer para convertir los mensajes de texto en vectores numéricos, representando la importancia de cada palabra en relación con el documento. Esta transformación es importante porque los modelos de aprendizaje automático no pueden trabajar con texto sin convertirlo a una representación numérica. Despues de usar TF IDF, la matriz tiene 5574 filas (textos) y 5000 columnas (características), lo que permite que el modelo de aprendizaje automático pueda procesar la información de manera efectiva.

```python
# Definimos los predictores (X) y la respuesta (y)
X = data['Processed_Message']  # Mensajes procesados
y = data['Class']  # Clases (spam o ham)

# Convertimos el texto a vectores de características utilizando TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)  # Limitamos a las 5000 características más importantes
X_tfidf = vectorizer.fit_transform(X).toarray()  # Convertimos a un array para compatibilidad con TensorFlow

print(X_tfidf.shape)  # Esto mostrará (número de documentos, número de características)
```

SMOTE (Synthetic Minority Over-sampling Technique) es una técnica que ayuda a equilibrar conjuntos de datos desbalanceados generando ejemplos sintéticos de la clase minoritaria, en este caso, el "spam". Al dividir el conjunto original en entrenamiento y prueba, SMOTE se aplica solo a los datos de entrenamiento, creando nuevas instancias basadas en la interpolación de ejemplos existentes de la clase minoritaria. Esto fue necesario porque el conjunto de datos desbalanceado que tenemos para este problema nos llevo a resultados muy desfaborables en un modelo enyrenado con este desbalance de datos obtuvimos valores de recall de 0.24 y f1 score de 0.39. En la matriz de confusio se observaron 122 falsos negativos y solo 39 verdaderos Negativos, siendo resultados malos porque practicamente era incapaz de detectar el spam. Al volver a entrenar el modelo con este nuevo conjunto de datos se obtuvieron resutlados muy buenos que se mostraran a continuacion.

```python
from imblearn.over_sampling import SMOTE  # Importamos SMOTE para el balanceo de clases

# División de los datos en conjunto de entrenamiento y de prueba
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Aplicamos SMOTE a los datos de entrenamiento para balancear las clases
smote = SMOTE(random_state=42)  # Inicializamos SMOTE
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)  # Ajustamos y re-muestreamos los datos

# Gráfico de conteo para visualizar la distribución utilizando SMOTE
sns.countplot(x=y_train_balanced)
plt.title("Class Distribution After SMOTE")
plt.xticks(ticks=[0, 1], labels=['Ham (0)', 'Spam (1)'])
plt.xlabel('Class')
plt.ylabel('Number of Instances')
plt.show()
```

El bloque de código se encarga de crear, compilar y entrenar un modelo de red neuronal de una solo neurona, para clasificar mensajes como spam o no spam. Primero, se define una función que construye un modelo secuencial con una capa de entrada y una neurona de salida que utiliza la función de activación sigmoide, adecuada para clasificación binaria. Luego, se compila el modelo usando el optimizador RMSprop, esto debido a que se probó con 5 diferentes optimizadores y los de mejores resultados fueron con Adam, Adamax y RMSprop, al final con RMSprop se obtenía un mejor recall y f1 score para los mensajes de spam, también obteniendo menor número de falsos negativos en nuestra matriz de confusión, llegando a la conclusión que era el mejor otpimizador para este problema de clasificación. La función de pérdida binary_crossentropy es la adecuada debido a que es un problema de clasificación y más específico un problema binario de si o no.

El entrenamiento se realiza mediante el método fit(), donde se especifican los datos, el número de épocas y el tamaño del lote. Esto ayuda a prevenir el sobreajuste y asegura que el modelo generalice bien a nuevos datos. Tras el entrenamiento, el modelo se evalúa con un conjunto de prueba para obtener métricas de desempeño y se guarda en un archivo.

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.optimizers import Adagrad
from tensorflow.keras.optimizers import Adamax


# Definición del modelo CNN con un solo neurona
def get_model():
    # Crea un modelo secuencial
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(input_shape=(X_train_balanced.shape[1],)),  # Capa de entrada
        tf.keras.layers.Dense(1, activation='sigmoid')  # Capa de salida con 1 neurona para clasificación binaria
    ])
    return model

# Compilar y entrenar el modelo
model = get_model()  # Llama a la función para obtener el modelo

# Establecer una tasa de aprendizaje específica
learning_rate = 0.01  # 0.1 / 0.01 / 0.001 (defalut) Cambia este valor según tus necesidades
optimizer = RMSprop(learning_rate=learning_rate)

model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Entrena el modelo
history=model.fit(X_train_balanced, y_train_balanced, epochs=20, batch_size=32, validation_split=0.2)

# Evaluar el modelo
loss, accuracy = model.evaluate(X_test, y_test)  # Evalúa el modelo usando el conjunto de prueba
print(f"Test Accuracy: {accuracy:.4f}")  # Imprime la precisión del modelo en el conjunto de prueba

# Guardar el modelo
model.save('spam_classifier_model.h5')  # Guarda el modelo entrenado en un archivo
```

A continuación se muestran dos gráficas una del Accuracy a través de las épocas de entrenamiento, así como de la pérdida. Se puede ver cómo en ambas gráficas ocurre lo esperado, el **Accuracy** mejora en cada época y la pérdida disminuye. De igual forma se alcanza a observar que existe una diferencia entre los valores finales con los datos de entrenamiento y los de validación, esto de igual forma es esperado debido a que los datos de validación son datos que el modelo no utilizo para entrenar, sin embargo la diferencia entre ambos es pequeña tanto en accuracy como en la pérdida, lo que indica buenos resultado de nuestro modelo y elimina la posibilidad de sobreajuste.

```python

# Visualización de resultados
plt.figure(figsize=(12, 5))  # Crea una figura para graficar la precisión y la pérdida

# Graficar la precisión
plt.subplot(1, 2, 1)  # Define la posición del gráfico
plt.plot(history.history['accuracy'], label='Train Accuracy')  # Graficar precisión de entrenamiento
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')  # Graficar precisión de validación
plt.title('Accuracy over Epochs')  # Título del gráfico
plt.xlabel('Epochs')  # Etiqueta del eje x
plt.ylabel('Accuracy')  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda

# Graficar la pérdida
plt.subplot(1, 2, 2)  # Define la posición del gráfico
plt.plot(history.history['loss'], label='Train Loss')  # Graficar pérdida de entrenamiento
plt.plot(history.history['val_loss'], label='Validation Loss')  # Graficar pérdida de validación
plt.title('Loss over Epochs')  # Título del gráfico
plt.xlabel('Epochs')  # Etiqueta del eje x
plt.ylabel('Loss')  # Etiqueta del eje y
plt.legend()  # Muestra la leyenda

plt.show()  # Muestra los gráficos

# Evaluar el modelo nuevamente
loss, accuracy = model.evaluate(X_test, y_test)  # Evalúa el modelo usando el conjunto de prueba
print(f"Test Accuracy: {accuracy:.4f}")  # Imprime la precisión del modelo en el conjunto de prueba

# Realizar predicciones usando el modelo entrenado
y_pred = (model.predict(X_test) > 0.5).astype(int)  # Predicciones binarizadas basadas en un umbral de 0.5
```

El modelo logró un rendimiento destacado en la clasificación de mensajes, obteniendo 945 verdaderos positivos (TP) y 149 verdaderos negativos (TN), lo que se traduce en una precisión general del 98%. En términos de métricas específicas, se logró una precisión del 99% para los mensajes ham (clase 0), mientras que para los mensajes spam (clase 1), la precisión fue del 94%. Sin embargo, el modelo mostró un recall del 99% para los mensajes ham y un 93% para los mensajes spam, indicando que aún existe un número notable de falsos negativos (12) en la clasificación de mensajes spam. A pesar de que, al ser un modelo de una sola neurona, se considera que tiene un rendimiento efectivo, es importante recordar que en situaciones del mundo real, incluso los sistemas de detección de spam en mensajes de texto o correo electrónico no son infalibles, y es común que algunos mensajes de spam se escapen a la bandeja de entrada principal.

```python
# 5. Compute Confusion Matrix and Performance Metrics
confusion_matrix = ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title("Confusion Matrix")
plt.show()

# Print classification report
print(classification_report(y_test, y_pred))

# 6. Save the model
model.save('spam_classifier_model.h5')
print("Model saved successfully.")
```

# Discusión y Conclusiones

Limitaciones de una red de una sola neurona en tareas complejas: Las redes neuronales con una sola neurona, como un perceptrón, son incapaces de modelar relaciones no lineales complejas en los datos. Este tipo de red puede clasificar bien problemas linealmente separables, pero su rendimiento disminuye significativamente en tareas que requieren múltiples capas de abstracción. La falta de complejidad en la arquitectura limita su capacidad para aprender patrones diferentes, lo que es común en aplicaciones de procesamiento de lenguaje natural y clasificación de texto, como en este caso con mensajes de spam.

Captura de la relación entre entradas y salidas: En este caso, el modelo de perceptrón mostró un rendimiento razonable, capturando algunas de las relaciones entre las características extraídas de los mensajes y su clasificación como spam o no spam. Sin embargo, la naturaleza simple del modelo implica que puede no haber capturado completamente todas las interacciones y matices presentes en los datos.

Velocidad de aprendizaje con diferentes tasas de aprendizaje: El modelo mostró una convergencia más rápida con tasas de aprendizaje más altas, aunque esto puede llevar a inestabilidad en el entrenamiento. Al probar con diferentes tazas, la mejor fue de 0.01, siendo de mayor velocidad y logrando mejor convergencia ya que también se obtuvieron mejores resultados para el modelo, con una mejor Accuracy y menor perdida, tanto en entrenamiento como validación.

Necesidad de normalizar/escala de las variables predictoras: Normalizar o escalar las variables predictoras es fundamental en redes neuronales para asegurar que todas las características tengan la misma escala, lo que ayuda al algoritmo a converger más rápidamente y evita que características con rangos de valores mucho mayores dominen el proceso de aprendizaje. En este caso, aunque no se aplicó una normalización explícita, sí fue necesario convertir el texto a valores numéricos y vectorizar los datos utilizando TF-IDF. Este proceso de vectorización es crucial, ya que permite al modelo interpretar los datos de texto como entradas numéricas, facilitando así su aprendizaje y análisis.

Estrategias para ajustar los hiperparámetros de entrenamiento: Se utilizaron diversas estrategias para ajustar los hiperparámetros, como la validación cruzada y la experimentación a prueba y error con diferentes tasas de aprendizaje, número de épocas y tamaños de lote. Estas técnicas permiten encontrar combinaciones de hiperparámetros que optimicen el rendimiento del modelo.

Convergencia del proceso de aprendizaje: En general, el proceso de aprendizaje convergió, aunque hubo momentos en que el modelo no alcanzó una convergencia óptima. Esto principalmente cuando modificamos el learning rate y probamos con diferentes optimizadores. Además, probablemente debido a la simplicidad de la arquitectura de una sola neurona limita su capacidad para adaptarse a patrones complejos en los datos.


Mejoras para resolver la aplicación de manera más eficiente y robusta: Una mejora clara sería adoptar una red neuronal de múltiples capas en lugar de un único perceptrón. Esto permitiría al modelo aprender representaciones más complejas de los datos, mejorando la capacidad de generalización y precisión en la clasificación de mensajes, especialmente en casos de datos desbalanceados como el spam. Limitaciones del Modelo: Aunque el modelo de una sola neurona logró buenos resultados, se evidenció que es limitado en tareas más complejas, lo que sugiere la necesidad de explorar arquitecturas más profundas o avanzadas en futuros proyectos.
